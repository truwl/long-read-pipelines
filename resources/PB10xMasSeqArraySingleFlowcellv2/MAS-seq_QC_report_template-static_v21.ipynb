```python
# Object to read in a bam stats file and parse it:
class BamStatsFile:
    """Interface to parsing and storing the results of `samtool stats`."""

    # CHK, Checksum [2]Read Names   [3]Sequences    [4]Qualities
    # Summary Numbers. Use `grep ^SN | cut -f 2-` to extract this part.
    # First Fragment Qualitites. Use `grep ^FFQ | cut -f 2-` to extract this part.
    # Last Fragment Qualitites. Use `grep ^LFQ | cut -f 2-` to extract this part.
    # GC Content of first fragments. Use `grep ^GCF | cut -f 2-` to extract this part.
    # GC Content of last fragments. Use `grep ^GCL | cut -f 2-` to extract this part.
    # ACGT content per cycle. Use `grep ^GCC | cut -f 2-` to extract this part. The columns are: cycle; A,C,G,T base counts as a percentage of all A/C/G/T bases [%]; and N and O counts as a percentage of all A/C/G/T bases [%]
    # Insert sizes. Use `grep ^IS | cut -f 2-` to extract this part. The columns are: insert size, pairs total, inward oriented pairs, outward oriented pairs, other pairs
    # Read lengths. Use `grep ^RL | cut -f 2-` to extract this part. The columns are: read length, count
    # Indel distribution. Use `grep ^ID | cut -f 2-` to extract this part. The columns are: length, number of insertions, number of deletions
    # Indels per cycle. Use `grep ^IC | cut -f 2-` to extract this part. The columns are: cycle, number of insertions (fwd), .. (rev) , number of deletions (fwd), .. (rev)
    # Coverage distribution. Use `grep ^COV | cut -f 2-` to extract this part.
    # GC-depth. Use `grep ^GCD | cut -f 2-` to extract this part. The columns are: GC%, unique sequence percentiles, 10th, 25th, 50th, 75th and 90th depth percentile
    
    class StatType(enum.Enum):

        def __new__(cls, *args, **kwds):
            value = len(cls.__members__) + 1
            obj = object.__new__(cls)
            obj._value_ = value
            return obj

        def __init__(self, delim_string, description):
            self.delim_string = delim_string
            self.description = description
            
        def __str__(self):
            return self.delim_string

        CHECKSUM               = "CHK", "Read Checksums"
        SUMMARY                = "SN", "Summary Statistics"
        FIRST_FRAG_QUAL        = "FFQ", "First Fragment Quality"
        LAST_FRAG_QUAL         = "LFQ", "Last Fragment Quality"
        GC_CONTENT_FIRST_FRAG  = "GCF", "GC Content First Fragment"
        GC_CONTENT_LAST_FRAG   = "GCL", "GC Content Last Fragment"
        AGCT_CONTENT_PER_CYCLE = "GCC", "AGCT Content Per Cycle"
        INSERT_SIZES           = "IS", "Insert Sizes"
        READ_LENGTHS           = "RL", "Read Lengths"
        INDEL_DISTRIBUTION     = "ID", "Indel Distribution"
        INDELS_PER_CYCLE       = "IC", "Indels Per Cycle"
        COVERAGE_DISTRIBUTION  = "COV", "Coverage Distribution"
        GC_DEPTH               = "GCD", "GC Depth"
        
        # Derived fields:
        READ_DENSITY           = "DERIVED_READ_DENSITY", "Read Density"
    
    _field_delim = '\t'
    
    def __init__(self,  file_name):
        self.file_name = file_name
        
        self.stats = dict()
        self.stats[BamStatsFile.StatType.SUMMARY] = dict()
        
        self._parse_file(file_name)
        
    def _parse_file(self, file_name):
        """Parse the data into friendly fields so we can work with it."""
        with open(file_name, 'r') as f:
            for line in f:
                # Handle comments:
                if line[0]  == "#":
                    continue
                
                # Handle special cases of stats:
                if line.startswith(str(BamStatsFile.StatType.CHECKSUM)):
                    self.stats[BamStatsFile.StatType.CHECKSUM] = line.split(self._field_delim)[1:]
                    
                elif line.startswith(str(BamStatsFile.StatType.SUMMARY)):
                    fields = line.strip().split(self._field_delim)
        
                    # Remove trailing ':'
                    k = fields[1][0:-1]
                    v = float(fields[2])

                    self.stats[BamStatsFile.StatType.SUMMARY][k] = v
                elif line.startswith(str(BamStatsFile.StatType.COVERAGE_DISTRIBUTION)):
                    fields = line.strip().split(self._field_delim)
                    
                    # Fix formatting of first field:
                    d = fields[1][1:-1].split('-') + fields[2:]
                    
                    # Fix '<' symbol and number of fields:
                    if d[0][-1] == '<':
                        d[0] = d[0][:-1]
                        d.insert(1, int(d[0]) ** 2)
                    
                    try:
                        self.stats[BamStatsFile.StatType.COVERAGE_DISTRIBUTION].append(d)
                    except KeyError:
                        self.stats[BamStatsFile.StatType.COVERAGE_DISTRIBUTION] = [d]
                else:
                    # Handle general stats here:
                    for stat in BamStatsFile.StatType:
                        if line.startswith(str(stat)):
                            try:
                                self.stats[stat].append(line[:-1].split(self._field_delim)[1:])
                            except KeyError:
                                self.stats[stat] = [line[:-1].split(self._field_delim)[1:]]
                
        # Convert simple fields into fast/numpy fields:
        for stat in BamStatsFile.StatType:           
            if stat != BamStatsFile.StatType.SUMMARY and stat != BamStatsFile.StatType.CHECKSUM:
                  
                try:
                    self.stats[stat] = np.asarray(self.stats[stat], dtype=np.float64, order='C')    
                except KeyError:
                    # This is OK - not all files have all stats.
                    pass   
        
        # Create data for read density:
        if BamStatsFile.StatType.READ_LENGTHS in self.stats:
            self.stats[BamStatsFile.StatType.READ_DENSITY] = np.copy(self.stats[BamStatsFile.StatType.READ_LENGTHS])
            self.stats[BamStatsFile.StatType.READ_DENSITY][:, 1] = self.stats[BamStatsFile.StatType.READ_DENSITY][:, 1] * self.stats[BamStatsFile.StatType.READ_DENSITY][:, 0]
            
    def display_all_stats_static(self):
        # Plot the Stats individually:
        
        pbar = tqdm(total=len(self.stats.items()), desc="Rendering Stat Plots")
        
        sq = int(np.ceil(np.sqrt(len(self.stats)-2)))
        fig, axs = plt.subplots(sq,sq)
        
        i = 0
        for stat, _ in self.stats.items():
            x = int(np.floor(i/sq))
            y = i % sq
            self.display_stat_static(stat, axs[x,y])
            
            if stat != BamStatsFile.StatType.SUMMARY and stat != BamStatsFile.StatType.CHECKSUM:
                i += 1
                
            pbar.update()
            pbar.refresh()
            
        fig.tight_layout(pad=2.0)
        plt.show()
        
        pbar.refresh()
        pbar.close()
                
    def display_stat_static(self, stat, axis=None):
        if stat in self.stats:
            do_show = False
            
            if stat == BamStatsFile.StatType.SUMMARY:
                # Create some plots of the information in the stats file:
                printmd(f"## {stat.description}")
                df = pd.DataFrame(self.stats[BamStatsFile.StatType.SUMMARY], index=[''])
                pretty_print_dataframe(df.T)
                
            elif stat == BamStatsFile.StatType.CHECKSUM:
                printmd(f"## {stat.description}")
                
                checksums = {
                    "Read Names": self.stats[stat][0],
                    "Sequences": self.stats[stat][1], 
                    "Qualities": self.stats[stat][2]
                }
                
                df = pd.DataFrame(checksums, index=[''])
                pretty_print_dataframe(df.T)
                
            elif stat == BamStatsFile.StatType.READ_LENGTHS:
                if axis is None:
                    do_show = True
                    _, axis = plt.subplots(1,1)
                    
                axis.plot(
                    self.stats[stat][:, 0],
                    self.stats[stat][:, 1]
                )
                axis.set_title(stat.description)
                axis.set_xlabel("Read Length")
                axis.set_ylabel("Number of Reads")
                
                if do_show:
                    plt.show()
                    
            elif stat == BamStatsFile.StatType.READ_DENSITY:
                    
                if axis is None:
                    do_show = True
                    _, axis = plt.subplots(1,1)
                    
                axis.plot(
                    self.stats[stat][:, 0],
                    self.stats[stat][:, 1] * self.stats[stat][:, 0]
                )
                axis.set_title(stat.description)
                axis.set_xlabel("Read Length")
                axis.set_ylabel("Density\n(Bases Read per Read Length)")
                
                if do_show:
                    plt.show()
            else:
                if axis is None:
                    _, axis = plt.subplots(1,1)
                    do_show = True
                axis.imshow(np.transpose(self.stats[stat]), aspect='auto')
                axis.set_title(stat.description)
                if do_show:
                    plt.show()
        else:
            print(f"Warning: BamStatsFile does not contain stat: {stat.delim_string} ({stat.description})", file=sys.stderr)
           
    def display_all_stats_interactive(self):
        # Plot the Stats individually:
        
        pbar = tqdm(total=len(self.stats.items()), desc="Rendering Stat Plots")
        
        sq = int(np.ceil(np.sqrt(len(self.stats)-2)))
        
        hv_graphs = []
        
        i = 0
        for stat, _ in self.stats.items():      
            if stat != BamStatsFile.StatType.SUMMARY and stat != BamStatsFile.StatType.CHECKSUM:
                hv_graphs.append(self.display_stat_interactive(stat, False))
                i += 1
                
            pbar.update()
            pbar.refresh()
            
        layout = hv.Layout(hv_graphs).cols(sq).opts(shared_axes=False)
        display(layout)
        
        pbar.refresh()
        pbar.close()
        
    def display_stat_interactive(self, stat, show=True):
        if stat in self.stats:            
            if stat == BamStatsFile.StatType.SUMMARY:
                # Create some plots of the information in the stats file:
                printmd(f"## {stat.description}")
                df = pd.DataFrame(self.stats[BamStatsFile.StatType.SUMMARY], index=[''])
                pretty_print_dataframe(df.T)
                
            elif stat == BamStatsFile.StatType.CHECKSUM:
                printmd(f"## {stat.description}")
                
                checksums = {
                    "Read Names": self.stats[stat][0],
                    "Sequences": self.stats[stat][1], 
                    "Qualities": self.stats[stat][2]
                }
                
                df = pd.DataFrame(checksums, index=[''])
                pretty_print_dataframe(df.T)
                
            elif stat == BamStatsFile.StatType.READ_LENGTHS:
                graph = hv.Curve(
                    (
                        self.stats[stat][:, 0],
                        self.stats[stat][:, 1]
                    )
                ).opts(
                    title=stat.description,
                    xlabel="Read Length",
                    ylabel="Number of Reads"
                )
                
                if show:
                    display(graph)
                    return None
                else:
                    return graph
                
            elif stat == BamStatsFile.StatType.READ_DENSITY:
                graph = hv.Curve(
                    (
                        self.stats[stat][:, 0],
                        self.stats[stat][:, 1]
                    )
                ).opts(
                    title=stat.description,
                    xlabel="Read Length",
                    ylabel="Density\n(Bases Read per Read Length)"
                )
                
                if show:
                    display(graph)
                    return None
                else:
                    return graph
            else:              
                shape = np.shape(self.stats[stat])
                    
                # Get base coords:
                x_coords = np.arange(shape[0])
                y_coords = np.arange(shape[1])
                
                # Use the hv.Image plotter here for speed:
                graph = hv.Image(
                    (
                        x_coords, y_coords, np.transpose(self.stats[stat])
                    ), 
                    datatype=['grid']
                ).opts(
                    title=stat.description, 
                    cmap="magma"
                )
                                
                if show:
                    display(graph)
                    return None
                else:
                    return graph

        else:
            print(f"Warning: BamStatsFile does not contain stat: {stat.delim_string} ({stat.description})", file=sys.stderr)
            

class ReadLengthMqData:
    
    def __init__(self, filename):
        self.filename = filename

        self.read_names = None
        self.mq_data = None
        self.read_lengths = None
        self.is_primary = None
        self.is_secondary = None
        self.is_supplementary = None
        
    def __str__(self):
        return f"{self.filename}"
    
    def __len__(self):
        return len(self.mq_data)

    def load_data(self, quiet=False, cache_only=False, do_cache=True, show_progress=True):
        data_suffix = ".full_npy_data"
        try:
            if do_cache:
                with open(self.filename + data_suffix, 'rb') as f:
                    if cache_only:
                        print(f"    Data already cached in: {self.filename}{data_suffix}")
                    else:
                        if not quiet:
                            print(f"    Loading data from cache: {self.filename}{data_suffix}")
                        self.mq_data = np.load(f, allow_pickle=True)
                        self.read_lengths = np.load(f, allow_pickle=True)
                        self.is_primary = np.load(f, allow_pickle=True)
                        self.is_secondary = np.load(f, allow_pickle=True)
                        self.is_supplementary = np.load(f, allow_pickle=True)
            else:
                raise FileNotFoundError("")
        except FileNotFoundError:
            with pysam.AlignmentFile(self.filename, 'r', check_sq=False) as alignment_file:

                if show_progress:
                    pbar = tqdm(desc=f"Ingesting reads")
                
                mapqs = []
                read_lengths = []
                read_names = []
                is_primary = []
                is_secondary = []
                is_supplementary = []
                for read in alignment_file.fetch(until_eof=True):
                    if read.is_unmapped:
                        mapq = -1
                    else:
                        mapq = read.mapping_quality

                    mapqs.append(mapq)
                    read_lengths.append(read.query_length)
                    read_names.append(read.query_name)
                    is_primary.append(not read.is_secondary and not read.is_supplementary)
                    is_secondary.append(read.is_secondary)
                    is_supplementary.append(read.is_supplementary)
                    
                    if show_progress:
                        pbar.update()

                if show_progress:
                    pbar.refresh()
                    pbar.close()
            
                self.read_names = np.array(read_names)
                self.mq_data = np.array(mapqs)
                self.read_lengths = np.array(read_lengths)
                self.is_primary = np.array(is_primary)
                self.is_secondary = np.array(is_secondary)
                self.is_supplementary = np.array(is_supplementary)

            # Save data to disk to make loading faster next time:
            if do_cache:
                with open(self.filename + data_suffix, 'wb') as f:
                    if not quiet:
                        print(f"    Saving data to cache: {self.filename}{data_suffix}")
                    np.save(f, self.read_names, allow_pickle=True)
                    np.save(f, self.mq_data, allow_pickle=True)
                    np.save(f, self.read_lengths, allow_pickle=True)
                    np.save(f, self.is_primary, allow_pickle=True)
                    np.save(f, self.is_secondary, allow_pickle=True)
                    np.save(f, self.is_supplementary, allow_pickle=True)
        
    def plot_mq_data(self):
        n, x, _ = plt.hist(self.mq_data, bins=np.arange(-2.5,61.5), histtype=u'step', density=False)
            
            
def read_ccs_metrics(ccs_report_file):
    """Current format is here: https://jbloomlab.github.io/alignparse/alignparse.ccs.html"""
    ccs_metrics = dict()
    
    with open(ccs_report_file, 'r') as f:
        for line in f:
            clean_line = line.strip()
            if len(clean_line) > 0:
                delim_indx = clean_line.find(':')
                field_name = clean_line[:delim_indx].strip()
                field_value = clean_line[delim_indx+1:].strip()
                if '(' in field_name:
                    field_name = field_name[:field_name.find('(')].strip()
                if '(' in field_value:
                    field_value = field_value[:field_value.find('(')].strip()  
                if len(field_value) != 0:
                    ccs_metrics[field_name] = int(float(field_value))
    return ccs_metrics

def extract_num_passes_from_ccs_bam(ccs_bam):
    with pysam.AlignmentFile(ccs_bam, 'rb', check_sq=False) as alignment_file:
        num_passes = []
        for read in alignment_file.fetch(until_eof=True):
            num_passes.append(read.get_tag("np"))
    return np.array(num_passes)

def extract_mapping_quality_from_bam(bam):
    with pysam.AlignmentFile(bam, 'rb', check_sq=False) as alignment_file:
        read_name_to_mq_dict = dict()
        for read in alignment_file.fetch(until_eof=True):
            read_name_to_mq_dict[read.query_name] = read.mapping_quality
    return read_name_to_mq_dict

class MarkerData:
    
#     _rc_delimiter = "_RC_"
    _rc_delimiter = "_RC"
    
    def __init__(self, name, start, end, qual):
        self.start = start
        self.end = end
        self.qual = qual
        self.is_rc = MarkerData._rc_delimiter in name
        
        if self.is_rc:
            self.name = name[:name.find(MarkerData._rc_delimiter)]
        else:
            self.name = name

def read_marker_alignments_file(marker_file):
    """
    Reads in the given file containing marker alignments from the 
    Extract Bounded Regions tool.
    :returns: dictionary of read_name -> list(MarkerData)
    """
    marker_dict = dict()
    with open(marker_file, 'r') as f:
        for line in f:
            fields = line.split("\t")
            read_name = fields[0]
            
            marker_list = []
            
            for field in fields[1:]:
                matches = re.match("(.*?):(\d+)-(\d+)@(\d+)", field)
                marker_list.append(
                    MarkerData(matches[1], int(matches[2]), int(matches[3]), int(matches[4]))
                )
                
            marker_dict[read_name] = marker_list
    return marker_dict
            

class SequenceData:
    def __init__(self, marker1, marker2):
        self.start_mark = marker1
        self.end_mark = marker2
        
        self.start = marker1.start
        self.end = marker2.end
        
        self.is_rc = marker1.is_rc
        
    def __str__(self):
        return f"[{self.start_mark.name},{self.start_mark.start},{self.start_mark.end},{self.start_mark.qual}]->[{self.end_mark.name},{self.end_mark.start},{self.end_mark.end},{self.end_mark.qual}]"
        
    def mas_seq_short_name(self, rc_dec="'"):
        if self.start_mark.name.startswith("oh"):
            return self.start_mark.name[2] + (rc_dec if self.is_rc else "")
        else:
            return self.end_mark.name[2] + (rc_dec if self.is_rc else "")
            
            
def get_sequence_string_from_seq_data_list(seq_data_list, is_bounded_regions=True, rc_dec="'"):
    """Gets a space-separated string of short overhang identifiers that represents
    the order of adapters found in the given list of SequenceData objects.
    """
    
    seq_data_list.sort(key = lambda s: s.start)
    
    # Old impl - good for bounded regions only.
    # that is, the splitting was done based on a unique start sequence and a 
    # universal terminator sequence.  Because of this we would only
    # have to grab the actual overhang / adapter name from one end of the 
    # read:
    if is_bounded_regions:
        return " ".join([f"{seq.mas_seq_short_name()}" for seq in seq_data_list])
    
    # OK, here we used naive splitting on delimiters.
    # so we have to do a little more work.
    
    seq_names_list = []
    for seq in seq_data_list:
        n = seq.start_mark.name
        if n.startswith("oh"):
            n = n[2]
            seq_names_list.append(n + (rc_dec if seq.is_rc else ""))
        else:
            seq_names_list.append(n)
        
    n = seq_data_list[-1].end_mark.name
    if n.startswith("oh"):
        n = n[2]
        seq_names_list.append(n + (rc_dec if seq.is_rc else ""))
    else:
        seq_names_list.append(n)
    
    return " ".join(seq_names_list)
        
        
def read_sequence_alignments_file(marker_file):
    """
    Reads in the given file containing sequence alignments from the 
    Extract Bounded Regions tool.
    :returns: dictionary of read_name -> list(SequenceData)
    """
    seq_dict = dict()
    with open(marker_file, 'r') as f:
        for line in f:
            fields = line.split("\t")
            read_name = fields[0]
            
            seq_list = []
            
            for field in fields[1:]:
                field = field[1:-1]
                markers = field.split("<>")
                m1 = re.match("(.*?):(\d+)-(\d+)@(\d+)", markers[0])
                m2 = re.match("(.*?):(\d+)-(\d+)@(\d+)", markers[1])
                seq_list.append(
                    SequenceData(
                        MarkerData(m1[1], int(m1[2]), int(m1[3]), int(m1[4])),
                        MarkerData(m2[1], int(m2[2]), int(m2[3]), int(m2[4]))
                    )   
                )
            seq_dict[read_name] = seq_list
    return seq_dict


def read_sequence_split_annotations_file(annotations_file):
    """
    Reads in the given file containing sequence alignments from the 
    Extract Bounded Regions tool.
    
    annotations_file is assumed to correspond to the annotated_read_sections_file
    from the split_sequences_on_delimiters.py tool with data in three tab-separated columns:
    
    READ_NAME READ_LENGTH SEGMENTS
    
    where SEGMENTS is of the form:
    
    <segment_name>[<segment_start>-<segment_end>]c<segment_cigar>@<segment_quality>,<segment_name>...
    
    :returns: dictionary of read_name -> list(SequenceData)
    """
    seq_dict = dict()
    with open(annotations_file, 'r') as f:
        is_first_line = True
        for line in f:
            
            # Skip the first (header) line:
            if is_first_line:
                is_first_line = False
                continue
                
            fields = line.split("\t")
            read_name = fields[0]
            read_length = int(fields[1])
            
            seq_list = []
            
            markers = fields[2].split(",")
            
            # TODO: For these coordinates make sure it's consistent with the alignments - is it 0 based?  Is it inclusive?
            last_marker_name = "START"
            last_marker_start = 0
            last_marker_end = 0
            last_marker_qual = 60
            
            for marker in markers:
                match = re.match("(.*?)\[(\d+)-(\d+)\]c(.*?)@(\d+.?\d*)", marker)
                marker_name = match[1]
                marker_start = int(match[2])
                marker_end = int(match[3])
                marker_cigar = match[4]
                marker_qual = float(match[5])
                
                seq_list.append(
                    SequenceData(
                        MarkerData(last_marker_name, last_marker_start, last_marker_end, last_marker_qual),
                        MarkerData(marker_name, marker_start, marker_end, marker_qual)
                    )
                )
                
                last_marker_name = marker_name
                last_marker_start = marker_start
                last_marker_start = marker_end
                last_marker_qual = marker_qual
            
            # add in the last marker delimiters:
            seq_list.append(
                SequenceData(
                    MarkerData(last_marker_name, last_marker_start, last_marker_end, last_marker_qual),
                    MarkerData("END", read_length, read_length, 60)
                )
            )
                
            seq_dict[read_name] = seq_list
    return seq_dict

def plot_read_sections(read_name, read_length, read_sequence_dict, seq_alpha=1):
    
    if read_name not in read_sequence_dict:
        raise KeyError(f"Read Name not in given read_sequence_dict: {read_name}")
    
    read_width = 1
    seq_width = read_width * .8
    
    x_coord = 10
    txt_offset = read_width/4 + read_width/2
    
    fig, axs = plt.subplots(1, 1, figsize=(gFIG_SIZE_in[0],gFIG_SIZE_in[1]))
    
    read_rect = matplotlib.patches.Rectangle(
        xy=(x_coord - read_width/2, 0), 
        width=read_width,
        height=read_length, 
        fill=True, 
        ec=None, 
        fc=[.5,.5,.5]
    )
    axs.add_patch(read_rect)
       
    colors = iter(plt.cm.rainbow(np.linspace(0,1,len(read_sequence_dict[read_name]))))
    for seq in read_sequence_dict[read_name]:
        
        c=next(colors)
        
        seq_rect = matplotlib.patches.Rectangle(
            xy=(x_coord - seq_width/2, seq.start), 
            width=seq_width, 
            height=seq.end - seq.start, 
            fill=True, 
            ec=None,
            fc=c, 
            alpha=seq_alpha
        )
        axs.add_patch(seq_rect)
        plt.text(x_coord + txt_offset,  seq.start + (seq.end - seq.start)/2, f"{seq.mas_seq_short_name('_rc')}")
    
    plt.xticks([x_coord], [read_name])
    axs.set_xlim(0, x_coord*2)
    axs.set_ylim(0, read_length * 1.1)
    fig.show()
    
def read_10x_metrics_file(file_name):
    ten_x_metrics = dict()
    ten_x_metric_names = []
    ten_x_metric_values = []
    with open(file_name, 'r') as f:
        tsv = csv.reader(f, delimiter='\t')
        header = next(tsv)
        values = next(tsv)
        for k,v in zip(header, values):
            ten_x_metrics[k] = int(float(v))

            ten_x_metric_names.append(k)
            ten_x_metric_values.append(int(float(v)))
    return ten_x_metrics, ten_x_metric_names, ten_x_metric_values



def read_rnaseq_metrics_file(file_name):
    rna_seq_metrics = dict()
    with open(file_name, 'r') as f:
        line = f.readline()
        while len(line) != 0:
            if (not line.startswith("#")) and (len(line.strip()) != 0):
                if line.startswith("PF_BASES"):
                    keys = line.strip().split("\t")
                    line = f.readline()
                    values = line.strip().split("\t")
                    for i in range(len(keys)):
                        try:           
                            if len(values[i]) == 0:
                                rna_seq_metrics[keys[i]] = 0
                            else:
                                rna_seq_metrics[keys[i]] = float(values[i])
                        except IndexError:
                            # Some of the metrics may not be there.
                            pass
                elif line.startswith("normalized_position"):
                    norm_pos = []
                    norm_cov = []
                    line = f.readline()
                    while (not line.startswith("#")) and (len(line.strip()) != 0):
                        pos, cov = line.strip().split("\t")
                        norm_pos.append(float(pos))
                        norm_cov.append(float(cov))
                        line = f.readline()
                    
                    norm_pos = np.array(norm_pos)
                    norm_cov = np.array(norm_cov)
                    
                    rna_seq_metrics['coverage_hist'] = (norm_pos, norm_cov)
                    # We've already read the next line, so we shouldn't do it below:
                    continue
                    
            line = f.readline()
            
    return rna_seq_metrics
```

## Ligation Heatmap
This is a chart indicating to which overhangs each MAS-seq adapter connects.  The hotter the cell, the more ligations occurred between those two overhang adapters.

```python
# Create a matrix with rows and columns corresponding to the possible elements in the array:
# MASSeq v1:
#     A B C D E F G H A' B' C' D' E' F' G' H'
# MASSeq v2:
#     A B C D E F G H I J K L M N O A' B' C' D' E' F' G' H' I' J' K' L' M' N' O' 

# The way we made this dataset with the original bounded region splitting
# uses the following construction for the heatmap:
if has_bounded_region_splits:

    with open(args.ebr_bounds_file, 'r') as f:
        max_num_elements = len(f.readlines())

    # Construct the heat-map matrix to accumulate counts:
    # Need 2x because of the RC direction:
    ligation_heat_matrix = np.zeros((max_num_elements*2, max_num_elements*2), dtype=int)
    index_map = dict()
    for k,v in {chr(ord('A')+i): i for i in range(max_num_elements)}.items():
        index_map[k] = v
    for k,v in {chr(ord('A')+i): i for i in range(max_num_elements)}.items():
        index_map[k+"'"] = v+max_num_elements

    # Calculate the counts for each element in the heat map:
    for profile in list(library_arrangement_count_dict.keys()):
        adapters = profile.split(" ")
        cur_adapter = adapters[0]
        for next_adapter in adapters[1:]:
            ligation_heat_matrix[index_map[cur_adapter]][index_map[next_adapter]] += 1
            cur_adapter = next_adapter

# Here we used the naive splitter method to separate the array elements
# because of this we need to account for the start / end of each element 
# in addition to the adapters.
# We will also likely notice that the first and last adapters are not 
# represented in the data.  This is because they were not included in the
# splitter delimiters since they are not needed (splitting on a leading
# delimiter would result in an empty sequence and the whole remaining
# sequence, and woulnd't add any value).
else:
    # The number of elements in the MASseq v2 library prep is fixed
    # and consists of overhangs A -> P
    max_num_elements = 16
    
    # Construct the heat-map matrix to accumulate counts:
    
    # Need 2x because of the RC direction, but we know that the 
    # first and last array elements won't be present because of 
    # how we did the splitting.
    # Therefore we need 2 fewer base rows / columns:
    max_num_elements = 16 - 2
    
    ligation_heat_matrix = np.zeros((2+max_num_elements*2, 2+max_num_elements*2), dtype=int)
    
    index_map = dict()
    index_map['START'] = 0
    # The indices here look messy because we added the START element above.
    # i.e. all the 1 offsets are because of this.
    for k,v in {chr(ord('B')+i-1): i for i in range(1, max_num_elements+1)}.items():
        index_map[k] = v
    for k,v in {chr(ord('B')+i-1): i for i in range(1, max_num_elements+1)}.items():
        index_map[k+"'"] = v+max_num_elements
    index_map['END'] = max_num_elements*2 + 1
    
    # Calculate the counts for each element in the heat map:
    for profile in list(library_arrangement_count_dict.keys()):
        adapters = profile.split(" ")
        cur_adapter = adapters[0]
        for next_adapter in adapters[1:]:
            adapter_1_ndx = index_map[cur_adapter]
            adapter_2_ndx = index_map[next_adapter]
            ligation_heat_matrix[adapter_1_ndx][adapter_2_ndx] += 1
            cur_adapter = next_adapter
```

# Raw Subread (Approxmiate Polymerase Read) Array Element Counts / Lengths

```python
array_element_mq_length_data = ReadLengthMqData(args.array_element_bam_file)
array_element_mq_length_data.load_data(do_cache=False)
# array_elements_and_mqs = extract_mapping_quality_from_bam(args.array_element_bam_file)
```

```python
# Create our data source to plot with bokeh

hist_values, bin_edges = np.histogram(array_element_mq_length_data.mq_data, bins=np.arange(-2,62), density=False)
primary_hist_values, _ = np.histogram(array_element_mq_length_data.mq_data[array_element_mq_length_data.is_primary], bins=np.arange(-2,62), density=False)

d_domain = bin_edges[:-1]
d_range = hist_values
num_vals = len(array_element_mq_length_data)
mean_val = np.average(d_domain, weights=d_range)
primary_mean_val = np.average(d_domain, weights=primary_hist_values)

handles = []
fig = plt.figure(figsize=gFIG_SIZE_in)
ax = fig.add_subplot()

width = 0.5

# Plot the bar graph:
h = ax.bar(
    d_domain, 
    d_range, 
    width,
    label="All", 
    align="center",
    color=[0, 0, .8]
)
handles.append(h)

mean_val

# Plot the bar graph:
h = ax.bar(
    d_domain + width, 
    primary_hist_values, 
    width,
    label="Primary Only", 
    align="center",
    color=[0, .8, 0]
)
handles.append(h)

primary_mean_val

# Plot the mean lines:
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
yrange = ymax-ymin

h, = ax.plot(
    [mean_val, mean_val], 
    [ymin-(yrange*.1), ymax+(yrange*.2)], 
    "--", 
    color=[0, 0, 0.4], 
    label=f"Mean Mapping Quality - All ({mean_val:2.02f})"
)
handles.append(h)

h, = ax.plot(
    [primary_mean_val, primary_mean_val], 
    [ymin-(yrange*.1), ymax+(yrange*.2)], 
    "--", 
    color=[0, .4, 0], 
    label=f"Mean Mapping Quality - Primary ({primary_mean_val:2.02f})"
)
handles.append(h)

ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax+(yrange*.1))

ax.set_title("Array Element Mapping Quality Counts")
ax.set_xlabel("Mapping Quality")
ax.set_ylabel("Number of Reads")

legend_anchor = (0.5, 0.2)
fig.legend(
    handles=handles, 
    title="Read Alignment", 
    bbox_to_anchor=legend_anchor, 
    loc="upper center", 
    fancybox=True, 
    shadow=True
)
fix_matplotlib_plot_visuals(fig, title_font_size, title_font_size, tick_label_font_size, tight_rect=[0,0.2,1,1])
fig.show()
```
